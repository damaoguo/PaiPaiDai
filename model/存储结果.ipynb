```python
import pandas as pd
import numpy as np
```

```python
# 复制行数,构造0-31共32个结果
test = pd.read_csv('../data/test.csv')

df =test[['listing_id', 'auditing_date', 'due_date','due_amt']]

newdf=pd.DataFrame()

# 复制行数(32行)

newdf = pd.DataFrame(np.repeat(df.values,32,axis=0))
newdf.columns = df.columns

day=[i for i in range(32)]
days=[]
for i in range(130000):
    days+=day
days=np.array(days)

newdf['days']=days

newdf['auditing_date']=pd.to_datetime(newdf['auditing_date'])
newdf['due_date']=pd.to_datetime(newdf['due_date'])

newdf['repay_date_0']=newdf['auditing_date']+pd.to_timedelta(newdf['days'],unit='d')
newdf['repay_date_1']=newdf['due_date']-pd.to_timedelta(newdf['days'],unit='d')

newdf.to_csv("../data/submission_template.csv",index=False)
```

# 2.合并二分类和多分类的结果

```python
sub_baseline_v8=pd.read_csv("../data/sub_baseline_v8.csv")
paipaidai_binary=pd.read_csv("../data/paipaidai_binary.csv")
```

```python
sub_baseline_v8.head(10)
```

```python
paipaidai_binary.head(10)
```

```python
sub_baseline_v8=sub_baseline_v8.merge(paipaidai_binary,on='listing_id',how='left')
```

```python
drop_left=sub_baseline_v8[sub_baseline_v8['binary']==1].drop_duplicates(subset='listing_id', keep='first')

sub_baseline_v8=sub_baseline_v8.drop(sub_baseline_v8[sub_baseline_v8['binary']==1].index)

drop_left['repay_date']=None
drop_left['repay_amt']=None

sub_example=pd.concat([sub_baseline_v8,drop_left],axis=0)

sub_example[['listing_id', 'repay_date', 'repay_amt']].to_csv('../data/submission.csv', index=False)
```
